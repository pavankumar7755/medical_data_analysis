```python
#  6.	How do you calculate the percentage of patients who did not show up for their appointment?

df=pd.DataFrame(mad)   # changing series to datframe for table access
```

```python
df.ndim               # two dimensional
```

```python
No_show_percentage=(df['No-show']=='Yes').mean()*100
print(f"No-show percentage :{No_show_percentage :2f}%")
```

```python
#  7.	How can you group the data by gender and count the no-shows in each group?

no_show_by_gender=df[df['No-show']=="Yes"].groupby('Gender').size()
no_show_by_gender.reset_index(name='No Count')
df
#print(f"Data by Gender and Count of No_show :{no_show_by_gender.reset_index(name='No Count')}")
```

```python
#  8.	How would you find the average age of patients who showed up vs. those who didnt?


avg_age_patients=df.groupby('No-show').Age.mean().astype(float).round(2)
avg_age_patients
```

```python
df.groupby('No-show').Age.mean().astype(float).round(2).reset_index()

#df.groupby('No-show')['Age'].mean().round(2).reset_index()
```

```python
##  9.	How can you count how many patients received SMS reminders but still didnt show up?

df['SMS']=df['SMS_received'].apply(lambda x:'yes' if x==0 else 'No')
df
```

```python
# sms_received =df.groupby('No-show')['SMS'].count()


count = df[(df['SMS_received'] == 1) & (df['No-show'] == 'Yes')].shape[0]
count
```

```python
## 10.	How would you determine the neighborhood with the highest no-show rate?
tt=df['Neighbourhood'].unique().tolist()
tt
```

```python
highest_Noshow_rate=df.groupby('Neighbourhood')['No-show'].apply(lambda x:(x=='Yes').mean()*100).round(2)

highest_Noshow_rate.sort_values(ascending=False).head(5).reset_index()
```

```python
#  11.	How would you convert ScheduledDay and AppointmentDay to datetime objects?

df['ScheduledDay']=pd.to_datetime(df['ScheduledDay'])
df['AppointmentDay']=pd.to_datetime(df['AppointmentDay'])
df.head(5)
                                    
```

```python
#  12.	How do you calculate the number of days between the scheduling date and appointment date?

df['Day_Difference']=(df['ScheduledDay']-df['AppointmentDay']).dt.days
df
```

```python
#  13.	How would you find the average scheduling lead time for no-shows vs. shows?

avg_schtime=df.groupby('No-show')['Day_Difference'].mean().round(1).reset_index()
avg_schtime
```

```python
#  14.	How can you filter appointments that were scheduled on the same day as the appointment?

#df
filter_apt=df[(df['ScheduledDay'].dt.date==df['AppointmentDay'].dt.date)&(df['Age']>=45)&(df['No-show']=='Yes')].reset_index()
filter_apt
```

```python
#  15.	How do you handle incorrect or outlier values in the Age column (e.g., negative values)?

#tt=df['Age'].tolist()
#tt

#df['Age'].head(10)

df['Age']=df['Age'].clip(lower=0)
df[df['Age']==0].head(2)
```

```python
#  16.	How would you rename columns to follow snake_case naming conventions (e.g., No-show to no_show)?

df.rename(columns={'No-show':'no_show'},inplace=True)
df
```

```python
#  17.	How would you convert the No-show column into a Boolean column?

df['no_show']=df['no_show'].apply(lambda x:True if x=='Yes' else False)
df
```

```python
#  19.	How would you prepare data to plot the no-show rate by age group?

#df['no_show']=df['no_show'].apply(lambda x:1 if x==True else 0)
age_group=mad.groupby('No-show').Age.mean().round(0)
age_group.reset_index()
```
